### Canny edge detection

We use OpenCV's canny edge detection function to highlight areas of strong gradients in the image.

The Canny edge detector uses two Sobel operators to detect horizontal and vertical lines. Taking the gradient along the x direction chooses edges closer to vertical.

```python
imageG = cv2.imread('/home/allen/SDCND/Lane_Line_Detection/test_images/solidWhiteCurve.jpg',0)
# Applying only the sobel operator along the x axis
sobelX = cv2.Sobel(imageG, cv2.CV_64F, 1, 0) # 1,0 signifies the x direction
lowThreshold = 60
highThreshold = 200
imageBinary = np.zeros_like(sobelX)
imageBinary[(sobelX > lowThreshold) & (sobelX < highThreshold)] = 255
plt.imshow(imageBinary, cmap='gray')
```

```python
lowThreshold = 75
highThreshold = 175
imageEdges = cv2.Canny(imageG, lowThreshold, highThreshold)
print(len(imageEdges.shape))
displayImage(imageEdges)
```

White pixels trace out the detected edges. Applying Gaussian blur to the image before Canny edge detection can improve your output images to reduce noise. The Gaussian blur uses a gaussian filter that has higher values in the middle and lower around to replicate a Gaussian function (mountaneous shape determined by $\sigma$). Kernel size (matrix filter) for the same is an odd number because the averaging is better. You center the averaging result on the required pixel rather than in between two pixels

```python
kernelSize = 5
imageBlurEdges = cv2.Canny(cv2.GaussianBlur(image,(kernelSize,kernelSize),0), lowThreshold, highThreshold)
displayImage(imageBlurEdges)
```

Note: The Canny function applies a smoothing function internally

### Masking a region

```python
def polyMask(img, vertices):
    if len(img.shape) == 3:
        mask = np.zeros_like(img)
        cv2.fillPoly(mask, vertices, [255,255,255]) # 255 bitwise_and 15 = True, oly way its False is when it is 0
        maskedImage = cv2.bitwise_and(img, mask)
    else:
        mask = np.zeros_like(img)
        cv2.fillPoly(mask, vertices, 255)
        maskedImage = cv2.bitwise_and(img, mask)
    return maskedImage
```

```python
imageMasked = polyMask(imageBlurEdges, np.array([[(100,540),(900, 540), (600,340), (400,340)]]))
displayImage(imageMasked)
```

### Hough Transformation

Previously, the image had edge points, small edges, circles, noisy edges(not straight ones) etc. The edge points (features of the image) vote. Most votes would mean a line in image sapce. How do we collect similar lines adjaent to each other? we use their slopes and intercept=> the line is to some extent unique within the region of collection. A point in the iamge space is a line in Hough space. Two points on the same lines in the iamge is the intersection of two lines in Hough space. Polar representation of Hough space results in curves. A point is now a sinusoid.

For each edge point calculate the H(d,$\theta$) where $d = x\cos \theta - y \sin \theta$ and $(x, y)$ corresponds to an edge point in image space. Increment H by one corresponding to that d. Now you know that in Hough space (polar), if curves intersect the corresponding points in image space lie on the same line. Now collect all these and plot them. More itnersections of curves in HS would mean more amount of votes.

Reduce the effect of noise on HT by smoothing the image beforehand

```python
def houghSpace(img):
    rho = 2             # distance resolution in pixels of the Hough grid
    theta = np.pi/180   # only one half of 360 is taken to account for -d
    threshold = 50      # minimum number of votes (intersections in Hough grid cell)
    minLineLength = 25  # minimum number of pixels making up a line
    maxLineGap = 125    # maximum gap in pixels between connectable line segments
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength, maxLineGap)
    lineImage = np.copy(img)*0
    lineImage = np.dstack([lineImage,lineImage,lineImage])
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(lineImage, (x1, y1), (x2, y2), (255,0,0), 2) 
    vertices = np.array([[(100,540),(900, 540), (600,340), (400,340)]])
    lineImage = polyMask(lineImage, vertices)
    assert lineImage.shape == image.shape
    return cv2.addWeighted(image, 0.8, lineImage, 1, 0) 
```

```python
imageHough = houghSpace(imageMasked)
displayImage(imageHough)
```

### Image pre-processing in one function

```python
def imagePP(img, vertices, lowThreshold=50, highThreshold=150, kernelSize = 5):
    """
    1. Grayscale conversion
    2. Gaussian blur
    3. Canny edge detection
    4. Image masking
    5. Hough lines
    """
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    imageBlurEdges = cv2.Canny(cv2.GaussianBlur(image,(kernelSize,kernelSize),0), lowThreshold, highThreshold)
    imageMasked = polyMask(imageBlurEdges, vertices)
    return houghSpace(imageMasked) #returns an array which would print array to screen
```

The LAB color space encodes all brightness in L channel and is not present in A and B.

```python
def imagePP2(img, vertices, kernelSize = 5):
    """
    1. BGR -> HLS to identify yellow and white lane lines
    2. Extract yellow and white lane lines 
    3. Merge the above into one image
    4. HLS -> RGB -> GRAY -> Gaussian blur -> Canny -> Hough Transform
    """
    vertices = np.array([[(100,540),(900, 540), (600,340), (400,340)]])
    imageMasked = polyMask(image, vertices)
    imageHLS = cv2.cvtColor(imageMasked, cv2.COLOR_BGR2HLS)
    
    # Extract yellow lane lines from the image
    lowYThreshold = np.array([10, 100, 0])
    highYThreshold = np.array([40, 200, 220])
    imageMaskHLSY = cv2.inRange(imageHLS, lowYThreshold, highYThreshold)
    imageMaskHLSY = np.dstack([imageMaskHLSY,imageMaskHLSY,imageMaskHLSY])
    imageHLSY = cv2.bitwise_and(imageHLS, imageMaskHLSY) # both need to have the same size to perform bitwise, use dstack
    
    # Extract white lane lines from the image
    lowWThreshold = np.array([100, 140, 0])
    highWThreshold = np.array([200, 255, 255])
    imageMaskHLSW = cv2.inRange(imageHLS, lowWThreshold, highWThreshold)
    imageMaskHLSW = np.dstack([imageMaskHLSW,imageMaskHLSW,imageMaskHLSW])
    imageHLSW = cv2.bitwise_and(imageHLS, imageMaskHLSW)
    
    # Merge the above two images
    imageHLSYW = cv2.bitwise_or(imageMaskHLSW, imageMaskHLSY)
    
    # Reproduce the same operations as seen in the imagePP function
    imageBGR = cv2.cvtColor(imageHLSYW, cv2.COLOR_HLS2BGR)
    imageGray = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2GRAY)
    imageCanny = cv2.Canny(cv2.GaussianBlur(imageGray,(kernelSize,kernelSize),0), lowThreshold, highThreshold)
    imageHS = houghSpace(imageCanny)
    plt.imshow(imageHS, cmap ='gray')
```

### Interpolation of fragmented lane lines

The `cv2.line` function outputs all the line segments present in the image. These lines are unqiue based on their end points. To separate the left lane line from the right we make use of the slope of these lines. The right lane line has a positive slope while the left lane line has a negative slope.
Each line has a unique slope (m) and y intercept (b), we take the average (or linear regression) of (m,b) to obtain average slope and average intercept.

```python
def houghSpace2(img):
    rho = 2             # distance resolution in pixels of the Hough grid
    theta = np.pi/180   # only one half of 360 is taken to account for -d
    threshold = 50      # minimum number of votes (intersections in Hough grid cell)
    minLineLength = 25  # minimum number of pixels making up a line
    maxLineGap = 125    # maximum gap in pixels between connectable line segments
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength, maxLineGap)
    lines = drawLine(lines)
    lineImage = np.copy(img)*0
    lineImage = np.dstack([lineImage,lineImage,lineImage])
    for x1,y1,x2,y2 in lines:
        cv2.line(lineImage, (x1, y1), (x2, y2), (255,0,0), 5) 
    vertices = np.array([[(100,540),(900, 540), (600,340), (400,340)]])
    lineImage = polyMask(lineImage, vertices)
    return lineImage
```

```python
def drawLine(lines):
    leftLine_m = []
    leftLine_b = []
    rightLine_m = []
    rightLine_b = []
    # better to segment left and right using centre of picture rather than slope because unnecessary points get selected
    for line in lines:
        for x1,y1,x2,y2 in line:
            #if abs(x1-x2) < 0.0001:    # Avoid division by 0 and slope = infinity
            slope = (y1-y2)/(x1-x2)
            intercept = y1 - (slope * x1)
            if slope < -0.1:                     
                leftLine_m.append(slope)         
                leftLine_b.append(intercept)     # nan occurs after numpy operates on empty list
            elif slope > 0.1:
                rightLine_m.append(slope)
                rightLine_b.append(intercept)
    if len(leftLine_m) > 0:                       # alwasy check for fail cases, what if list was empty, division by zero
        leftSlope = np.mean(np.array(leftLine_m))
        leftIntercept = np.mean(np.array(leftLine_b))
        leftBotY = 540 # or image.shape[1]
        leftBotX = int((leftBotY - leftIntercept)/leftSlope)
        leftTopY = 340
        leftTopX = int((leftTopY - leftIntercept)/leftSlope)
    else:
        leftBotY = 0
        leftBotX = 0
        leftTopY = 0
        leftTopX = 0
        
    if len(rightLine_m) > 0:
        rightSlope = np.mean(np.array(rightLine_m))
        rightIntercept = np.mean(np.array(rightLine_b))
        rightBotY = 540 # or image.shape[1]
        rightBotX = int((rightBotY - rightIntercept)/rightSlope)
        rightTopY = 340
        rightTopX = int((rightTopY - rightIntercept)/rightSlope)
    else:
        rightBotY = 0
        rightBotX = 0
        rightTopY = 0
        rightTopX = 0
    
    return [[leftBotX,leftBotY,leftTopX,leftTopY],[rightBotX,rightBotY,rightTopX,rightTopY]]
```

```python

def imagePP3(img, vertices = np.array([[(100,540),(900, 540), (600,340), (400,340)]]), kernelSize = 5):
    """
    1. BGR -> HLS to identify yellow and white lane lines
    2. Extract yellow and white lane lines 
    3. Merge the above into one image
    4. HLS -> RGB -> GRAY -> Gaussian blur -> Canny -> Hough Transform
    """
    vertices = np.array([[(100,540),(900, 540), (600,340), (400,340)]])
    imageMasked = polyMask(img, vertices)   #Python could use global image file so use img
    imageHLS = cv2.cvtColor(imageMasked, cv2.COLOR_BGR2HLS)
    
    # Extract yellow lane lines from the image
    lowYThreshold = np.array([10, 100, 0])
    highYThreshold = np.array([40, 200, 220])
    imageMaskHLSY = cv2.inRange(imageHLS, lowYThreshold, highYThreshold)
    imageMaskHLSY = np.dstack([imageMaskHLSY,imageMaskHLSY,imageMaskHLSY])
    imageHLSY = cv2.bitwise_and(imageHLS, imageMaskHLSY) # both need to have the same size to perform bitwise, use dstack
    
    # Extract white lane lines from the image
    lowWThreshold = np.array([100, 140, 0])
    highWThreshold = np.array([200, 255, 255])
    imageMaskHLSW = cv2.inRange(imageHLS, lowWThreshold, highWThreshold)
    imageMaskHLSW = np.dstack([imageMaskHLSW,imageMaskHLSW,imageMaskHLSW])
    imageHLSW = cv2.bitwise_and(imageHLS, imageMaskHLSW)
    
    # Merge the above two images
    imageHLSYW = cv2.bitwise_or(imageMaskHLSW, imageMaskHLSY)
    
    # Reproduce the same operations as seen in the imagePP function
    imageBGR = cv2.cvtColor(imageHLSYW, cv2.COLOR_HLS2BGR)
    imageGray = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2GRAY)
    imageCanny = cv2.Canny(cv2.GaussianBlur(imageGray,(kernelSize,kernelSize),0), 100, 200)
    #global count
    #count = count+1
    #text = 'image'+str(count)+'.png'
    #cv2.imwrite(text, imageCanny)
    imageHS = houghSpace2(imageCanny)
     
    return cv2.addWeighted(img, 0.8, imageHS, 1, 0)
```
